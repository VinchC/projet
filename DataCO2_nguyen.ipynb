```python
df_global = df[df["country"] == "World"].copy()
```

```python
df_global = df_global[df_global["year"] >= 1880]
```

```python
# Forward fill theo thời gian
df_global = df_global.sort_values("year")
df_global = df_global.ffill()
```

```python
df_global.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df_global["year"], df_global["co2"])
plt.xlabel("Year")
plt.ylabel("CO₂ emissions")
plt.title("Global CO₂ emissions since 1880")
plt.show()
```

```python
plt.figure()
plt.plot(df_global["year"], df_global["temperature_change_from_ghg"])
plt.xlabel("Year")
plt.ylabel("Temperature change (°C)")
plt.title("Global temperature change since 1880")
plt.show()
```

```python
df_corr = df_global.select_dtypes(include="number")
df_corr.corr()

```

```python
import statsmodels.api as sm

X = df_global["total_ghg"]
y = df_global["temperature_change_from_ghg"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```


The analysis focuses on global-level data from 1880 onwards, a period for which temperature observations are considered reliable. Data cleaning was performed by restricting the dataset to global aggregates and removing years with incomplete observations. Forward filling was applied cautiously to preserve temporal continuity.


```python
numeric_df = df_global.select_dtypes(include=["float64", "int64"])
corr_matrix = numeric_df.corr()
corr_matrix
```

Correlation analysis was conducted using numerical variables only. Non-numeric columns were excluded prior to computation.

```python
df_trend = df_global.copy()

for col in ["co2", "methane", "total_ghg", "temperature_change_from_ghg"]:
    df_trend[col] = df_trend[col] / df_trend[col].iloc[0] * 100
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))
plt.plot(df_trend["year"], df_trend["co2"], label="CO2")
plt.plot(df_trend["year"], df_trend["methane"], label="Methane")
plt.plot(df_trend["year"], df_trend["total_ghg"], label="Total GHG")
plt.plot(df_trend["year"], df_trend["temperature_change_from_ghg"], label="Temperature")

plt.legend()
plt.xlabel("Year")
plt.ylabel("Index (1880 = 100)")
plt.title("Normalized trends of GHG emissions and temperature since 1880")
plt.show()
```

```python
df_diff = df_global.copy()

for col in ["co2", "total_ghg", "temperature_change_from_ghg"]:
    df_diff[col] = df_diff[col].diff()

df_diff = df_diff.dropna()
```

```python
plt.figure()
plt.scatter(
    df_diff["total_ghg"],
    df_diff["temperature_change_from_ghg"]
)
plt.xlabel("Δ Total GHG")
plt.ylabel("Δ Temperature")
plt.title("Year-to-year changes in GHG vs temperature")
plt.show()
```

```python
import statsmodels.api as sm

X = df_global[["total_ghg", "year"]]
y = df_global["temperature_change_from_ghg"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
df_ma = df_global.copy()
df_ma["temp_ma"] = df_ma["temperature_change_from_ghg"].rolling(10).mean()
df_ma["ghg_ma"] = df_ma["total_ghg"].rolling(10).mean()

plt.figure()
plt.plot(df_ma["year"], df_ma["temp_ma"], label="Temperature (10y MA)")
plt.plot(df_ma["year"], df_ma["ghg_ma"], label="GHG (10y MA)")
plt.legend()
plt.show()
```

```python
import pandas as pd

# 1. Charger le fichier CSV (adapter le chemin si nécessaire)
df = pd.read_csv('./data/owid-co2-data.csv')

```

```python
# 2. Liste des colonnes à conserver
columns_to_keep = [
    "year",
    "country",
    "iso_code",
    "population",
    "co2",
    "methane",
    "nitrous_oxide",
    "total_ghg",
    "temperature_change_from_ghg",
    "cumulative_co2",
    "co2_per_capita",
    "co2_growth_abs",
    "co2_growth_prct",
    "share_global_co2",
    "gdp",
    "energy_per_capita"
]

```

```python
# 3. Garder uniquement ces colonnes
df_clean = df[columns_to_keep]


```

```python
# 4. Filtrer les données à partir de 1850
df_clean = df_clean[df_clean["year"] >= 1850]

# 5. (Optionnel mais recommandé) Réinitialiser l’index
df_clean = df_clean.reset_index(drop=True)

# 6. Sauvegarder le nouveau fichier nettoyé
df_clean.to_csv("owid-co2-clean-1850.csv", index=False)

print("Nettoyage terminé. Fichier sauvegardé : owid-co2-clean-1850.csv")
```

Top 10 des pays émetteurs de CO₂

```python
# 2) Année la plus récente disponible
annee_max = int(df["year"].max())

# 3) Garder uniquement les pays (iso_code à 3 lettres) et prendre le Top 10
df_pays = df[df["iso_code"].astype(str).str.len() == 3]

top10 = (
    df_pays[df_pays["year"] == annee_max]
    .sort_values("co2", ascending=False)
    .head(10)
    .copy()
)

# 4) Préparer les étiquettes en français
# OWID co2 est généralement en "millions de tonnes" (Mt). On l'affiche tel quel.
top10["emissions"] = top10["co2"]

# 5) Tracer (barres horizontales, plus lisible)
plt.figure(figsize=(10, 6))
plt.barh(top10["country"][::-1], top10["emissions"][::-1])

plt.title(f"Top 10 des pays émetteurs de CO₂ (année {annee_max})")
plt.xlabel("Émissions de CO₂ (Mt)")
plt.ylabel("Pays")

# Optionnel: afficher les valeurs au bout des barres
for i, v in enumerate(top10["emissions"][::-1]):
    plt.text(v, i, f"  {v:.0f}", va="center")

plt.tight_layout()
plt.show()
```
