Nettoyage des données

```python
df.isna().sum()
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
df_global = df[df["country"] == "World"].copy()
```

```python
df_global = df_global[df_global["year"] >= 1880]
```

```python
# 2. Liste des colonnes à conserver
columns_to_keep = [
    "year",
    "country",
    "iso_code",
    "population",
    "co2", 
    "total_ghg",
    "temperature_change_from_co2",
    "cumulative_co2",
    "co2_per_capita",
     "gdp",
    "energy_per_capita"
]
```

```python
# 3. Garder uniquement ces colonnes
df_clean = df[columns_to_keep]
```

```python
# 4. Filtrer les données à partir de 1850
df_clean = df_clean[df_clean["year"] >= 1850]

# 5. (Optionnel mais recommandé) Réinitialiser l’index
df_clean = df_clean.reset_index(drop=True)

# 6. Sauvegarder le nouveau fichier nettoyé
df_clean.to_csv("owid-co2-clean-1850.csv", index=False)

print("Nettoyage terminé. Fichier sauvegardé : owid-co2-clean-1850.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df_global["year"], df_global["co2"])
plt.xlabel("Year")
plt.ylabel("CO₂ emissions")
plt.title("Global CO₂ emissions since 1880")
#Afficher
plt.tight_layout() 
plt.savefig('./images/Global CO2 emissions since 1880.png', dpi=300, bbox_inches='tight') 
plt.show()
```

```python
# Données pour la régression linéaire
x = df_global["year"].values
y = df_global["temperature_change_from_ghg"].values

# Linear regression: y = a*x + b
a, b = np.polyfit(x, y, 1)

# Slope suivre °C / decade
slope_decade = a * 10

# Créer les valeurs de la tendance linéaire
trend = a * x + b

plt.figure()
plt.plot(x, y, label="Observed data")
plt.plot(x, trend, '--', label="Linear trend")

plt.xlabel("Year")
plt.ylabel("Temperature change (°C)")
plt.title("Global temperature change since 1880")

# Afficher l'équation de la droite et la pente
eq_text = (
    f"y = {a:.4f}x + {b:.2f}\n"
    f"Slope = {slope_decade:.3f} °C / decade"
)

plt.text(
    0.02, 0.95, eq_text,
    transform=plt.gca().transAxes,
    fontsize=10,
    verticalalignment="top",
    bbox=dict(boxstyle="round", alpha=0.3)
)

plt.legend()
plt.tight_layout()
plt.savefig(  './images/Global temperature change since 1880.png',  dpi=300, bbox_inches='tight')
plt.show()
```

Report IPCC

CO₂ ↔ GDP (logique economic – Climate)

CO₂ per capita ↔ Energy per capita

```python

df_energy = df.dropna(
    subset=["energy_per_capita", "co2_per_capita", "gdp"]
)

df_energy["energy_group"] = pd.qcut(
    df_energy["energy_per_capita"],
    q=5,
    labels=[
        "Très faible",
        "Faible",
        "Moyen",
        "Élevé",
        "Très élevé"
    ]
)
```

```python
df_bar = (
    df_energy
    .groupby("energy_group", observed=True)
    .agg(
        co2_per_capita_mean=("co2_per_capita", "mean"),
        gdp_mean=("gdp", "mean")
    )
    .reset_index()
)
```

```python
plt.figure(figsize=(8,5))
plt.bar(
    df_bar["energy_group"],
    df_bar["co2_per_capita_mean"]
)
plt.xlabel("Niveau de consommation d’énergie par habitant")
plt.ylabel("CO₂ par habitant (tonnes)")
plt.title("CO₂ par habitant selon le niveau de consommation d’énergie")
#Afficher
plt.tight_layout() 
plt.savefig('./images/CO2 par habitant selon le niveau de consommation d\'énergie.png', dpi=300, bbox_inches='tight') 
plt.show()

```

Top 10 des pays émetteurs de CO₂

```python
# 2) Année la plus récente disponible
annee_max = int(df["year"].max())

# 3) Garder uniquement les pays (iso_code à 3 lettres) et prendre le Top 10
df_pays = df[df["iso_code"].astype(str).str.len() == 3]

top10 = (
    df_pays[df_pays["year"] == annee_max]
    .sort_values("co2", ascending=False)
    .head(10)
    .copy()
)

# 4) Préparer les étiquettes en français
# OWID co2 est généralement en "millions de tonnes" (Mt). On l'affiche tel quel.
top10["emissions"] = top10["co2"]

# 5) Tracer (barres horizontales, plus lisible)
plt.figure(figsize=(10, 6))
plt.barh(top10["country"][::-1], top10["emissions"][::-1])

plt.title(f"Top 10 des pays émetteurs de CO₂ (année {annee_max})")
plt.xlabel("Émissions de CO₂ (Mt)")
plt.ylabel("Pays")

# Optionnel: afficher les valeurs au bout des barres
for i, v in enumerate(top10["emissions"][::-1]):
    plt.text(v, i, f"  {v:.0f}", va="center")

#Afficher
plt.tight_layout() 
plt.savefig('./images/Top 10 des pays émetteurs de CO₂.png', dpi=300, bbox_inches='tight') 
plt.show()
```
