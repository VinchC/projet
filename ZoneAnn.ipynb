```python

df.columns = [
    "Year", "Global", "NH", "SH",
    "24N_90N", "24S_24N", "90S_24S",
    "64N_90N", "44N_64N", "24N_44N",
    "EQU_24N", "24S_EQU", "44S_24S",
    "64S_44S", "90S_64S"
]
df.info()
df.head()
df.describe()
```

```python
df["Year"] = df["Year"].astype(int)

for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

```

```python
na_col = df.isna().mean().sort_values(ascending=False)
na_col.head(20)
```

```python
df = df.set_index("Year")
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
col = num_cols[0]

label_name = "Zonal annual means"

df[col].plot(figsize=(10,4))
plt.title(f"Time series: {label_name}")
plt.xlabel("Year")
plt.ylabel(label_name)
plt.show()

```

```python
df["mean_all"] = df[num_cols].mean(axis=1)
df["mean_all"].plot(figsize=(10,4))
plt.title("Mean across numeric columns (all zones)")
plt.xlabel("Year")
plt.ylabel("Mean")
plt.show()
```

```python
col = num_cols[0]

df[col].plot(
    alpha=0.5,
    figsize=(10,4),
    label="Original data"
)

df[col].rolling(5).mean().plot(
    label="Rolling mean (5 years)"
)

df[col].rolling(10).mean().plot(
    label="Rolling mean (10 years)"
)

plt.title(f"{col} with rolling mean (5-year and 10-year windows)")
plt.xlabel("Year")
plt.ylabel("Temperature anomaly (°C)")
plt.legend()
plt.show()
```

```python
Z = (df[num_cols] - df[num_cols].mean()) / df[num_cols].std()
Z.mean(axis=1).plot(figsize=(10,4))
plt.title("Standardized mean (z-score) across zones")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df[num_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation heatmap")
plt.show()
```

```python
df_clean = df.copy()
df_clean[num_cols] = df_clean[num_cols].interpolate(method="linear", limit_direction="both")
```

```python
import pandas as pd, numpy as np, os
df = pd.read_csv("./data/ZonAnn.Ts+dSST.csv", sep=",", header=0, low_memory=False)
print(df.shape)
print(df.columns.tolist()[:15])
df.head(3)
```

```python
cols = ["Glob", "NHem", "SHem"]

df.set_index("Year")[cols].plot(figsize=(10,4))
plt.title("Évolution annuelle des anomalies de température")
plt.ylabel("Anomalie de température (°C)")
plt.xlabel("Année")
plt.legend(["Globale", "Hémisphère Nord", "Hémisphère Sud"])
plt.show()
```

NH montre plus vit que SH après 1980. 

```python
zones = ["64N-90N", "44N-64N", "24N-44N", "24S-24N"]

df.set_index("Year")[zones].plot(figsize=(10,4))
plt.title("Anomalies par zones latitudinales")
plt.ylabel("Anomalie de température (°C)")
plt.xlabel("Année")
plt.legend(zones)
plt.show()
```

64°N et 90°N (Arctic amplification): Les pays concernés sont : Canada (territoires arctiques), Russie (Sibérie arctique), Danemark via le Groenland, Norvège (nord du pays, Svalbard), États-Unis (Alaska), Suède (au nord du cercle polaire),Finlande (Laponie)
44°N et 64°N: Pays concernés (principaux): Europe, Asie, Amérique du Nord

```python
df_rm = df.set_index("Year").rolling(window=10).mean()

df_rm["Glob"].plot(figsize=(10,4))
plt.title("Anomalie globale – moyenne mobile 10 ans")
plt.ylabel("Anomalie de température (°C)")
plt.xlabel("Année")

plt.text(
    0.02, 0.95,
    "La moyenne mobile met en évidence\n"
    "une tendance croissante nette\n"
    "depuis la fin du XXe siècle.",
    transform=plt.gca().transAxes,
    fontsize=10,
    verticalalignment="top"
)

plt.show()
```
