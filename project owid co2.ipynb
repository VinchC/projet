```python
# Préparation du df

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import statsmodels.api as sm
import seaborn as sns

pd.set_option('display.max_columns', None)
df = pd.read_csv(filepath_or_buffer = './data/owid-co2-data.csv', sep = ',', header = 0, low_memory=False)
# df = pd.read_csv(filepath_or_buffer = './data/owid-co2-data.csv', sep = ',', header = 0, low_memory=False)
```

```python
# En pourcentage
df.isna().mean() * 100
```

```python
# Pourcentage de valeurs manquantes par année hors groupes (iso non vide)
missing_pct = (
    df
    .dropna(subset=["iso_code"])
    .groupby("year")
    .apply(lambda x: x.drop(columns=["year"]).isna().mean() * 100)
    .round(2)
)

missing_pct

missing_pct.round(2)

```

```python
# Version longue
missing_long = (
    missing_pct
    .reset_index()
    .melt(
        id_vars="year",
        var_name="variable",
        value_name="missing_pct"
    )
)

missing_long.sort_values("missing_pct", ascending=False)

missing_long
```

```python
# Heatmap de complétude des données manquantes par année et par variable
plt.figure(figsize=(14, 8))

sns.heatmap(
    missing_pct,
    cmap="Reds",
    annot=False,
    linewidths=0.5,
    cbar_kws={"label": "% valeurs manquantes"}
)

plt.title("Pourcentage de valeurs manquantes par année et par variable")
plt.xlabel("Variables")
plt.ylabel("Year")
plt.tight_layout()
plt.show()
```

```python
# Pourcentage de valeurs manquantes par année pour les groupes (iso vide)

missing_pct = (
    df
    .loc[df["iso_code"].isna()]
    .groupby("year")
    .apply(lambda x: x.drop(columns=["year"]).isna().mean() * 100)
    .round(2)
)

missing_pct
```

```python
missing_long = (
    missing_pct
    .reset_index()
    .melt(
        id_vars="year",
        var_name="variable",
        value_name="missing_pct"
    )
)

missing_long.sort_values("missing_pct", ascending=False)

missing_long
```

```python
# Heatmap de complétude des données manquantes par année et par variable pour les groupes (iso vide)

plt.figure(figsize=(14, 8))

sns.heatmap(
    missing_pct,
    cmap="Reds",
    linewidths=0.5,
    cbar_kws={"label": "% valeurs manquantes"}
)

plt.title("Valeurs manquantes (%) – lignes avec iso_code nul")
plt.xlabel("Variables")
plt.ylabel("Year")
plt.tight_layout()
plt.show()
```

```python
# Liste des groupes
countries_iso_null_df = (
    df
    .loc[df["iso_code"].isna(), ["country"]]
    .drop_duplicates()
    .sort_values("country")
    .reset_index(drop=True)
)

countries_iso_null_df
```

```python

countries = [
    "International aviation",
    "International shipping",
    "Kuwaiti Oil Fires",
    "Kuwaiti Oil Fires (GCP)"
]

# Filtrer les pays ciblés
df_filtered = df[df["country"].isin(countries)].copy()

# Colonnes à inclure dans le calcul de complétude
cols_to_check = df_filtered.columns.difference(["year", "country", "iso_code"])

# Complétude par ligne
df_filtered["completeness"] = (
    100 - df_filtered[cols_to_check].isna().mean(axis=1) * 100
)

# Agrégation par année et par country
completeness_year = (
    df_filtered
    .groupby(["year", "country"])["completeness"]
    .mean()
    .reset_index()
)

plt.figure()

for country in countries:
    subset = completeness_year[completeness_year["country"] == country]
    plt.plot(subset["year"], subset["completeness"], label=country)

plt.xlabel("Year")
plt.ylabel("Data completeness (%)")
plt.title("Data completeness by year – selected countries")
plt.legend()
plt.tight_layout()
plt.show()
```
