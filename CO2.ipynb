```python
# 1. Identifier toutes les colonnes "object" (sauf country et iso_code)
colonnes_a_convertir = df.select_dtypes(include=['object']).columns.tolist()

# Retirer les colonnes textuelles
colonnes_a_convertir = [col for col in colonnes_a_convertir if col not in ['country', 'iso_code']]

print(f"Colonnes à convertir : {colonnes_a_convertir}")

# 2. Nettoyer et convertir
for col in colonnes_a_convertir:
    df[col] = df[col].astype(str).str.strip()  # Supprimer espaces
    df[col] = df[col].replace(['', 'NaN', 'nan', 'None', '***'], pd.NA)  # Remplacer valeurs manquantes
    
    # Ajouter le 0 devant si nécessaire (comme avant)
    df[col] = df[col].str.replace(r'^\.', '0.', regex=True)  # .34 → 0.34
    df[col] = df[col].str.replace(r'^-\.', '-0.', regex=True)  # -.13 → -0.13
    df[col] = df[col].str.replace(r'^\+\.', '0.', regex=True)  # +.56 → 0.56
    
    # Convertir en float
    df[col] = pd.to_numeric(df[col], errors='coerce')


```

```python
df.dtypes
```

```python
#les colonnes à garder en texte
colonnes_texte = ['country', 'iso_code']

#Convertir toutes les autres colonnes en float
for col in df.columns:
    if col not in colonnes_texte:
        # Convertir en float, remplacer les erreurs par NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# 3. Vérifier
print(" Types après conversion :")
print(df.dtypes)
print(f"\n Shape : {df.shape}")
print("\n" + "="*50)
df.head()
```

```python
# Périodes clés (identiques à votre code initial)
periodes_cles = {
    '1880-1920': (1880, 1920, '#4A90E2'),
    '1921-1950': (1921, 1950, '#06A77D'),
    '1951-1976': (1951, 1976, '#F77F00'),
    '1977-2000': (1977, 2000, '#E63946'),
    '2001-2024': (2001, 2024, '#8B0000')
}

# Créer une colonne 'periode' dans le DataFrame
def assigner_periode(year):
    for periode, (start, end, _) in periodes_cles.items():
        if start <= year <= end:
            return periode
    return None  # Pour les années hors périodes (ex: avant 1880)

df['periode'] = df['year'].apply(assigner_periode)

# Vérifier la répartition
print("Répartition par période :")
print(df['periode'].value_counts())



```

```python
# Émissions globales par période (comme dans votre DataFrame global)
emissions_globales = (
    df.groupby('periode')['co2']
    .sum()
    .reset_index()
    .rename(columns={'co2': 'emissions_Mt'})
)

# Émissions par hémisphère et période
emissions_par_hemisphere = (
    df.groupby(['periode', 'hemisphere'])['co2']
    .sum()
    .unstack()
    .reset_index()
)

# Calcul des parts (%)
emissions_par_hemisphere['total'] = emissions_par_hemisphere['Nord'] + emissions_par_hemisphere['Sud']
emissions_par_hemisphere['part_Nord'] = (emissions_par_hemisphere['Nord'] / emissions_par_hemisphere['total']) * 100
emissions_par_hemisphere['part_Sud'] = (emissions_par_hemisphere['Sud'] / emissions_par_hemisphere['total']) * 100

# Fusion avec les émissions globales
resultats = emissions_globales.merge(emissions_par_hemisphere, on='periode')

print("\nÉmissions de CO₂ par période et hémisphère :")
display(resultats)



```
