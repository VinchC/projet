```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Préparation des données
df_filtered = df[(df['year'] >= 1880) & (df['year'] <= 2024)].copy()

#Calcul des émissions cumulées par période
periods = {
    '1880-1951': df_filtered[(df_filtered['year'] >= 1880) & (df_filtered['year'] <= 1951)]['co2'].sum(),
    '1951-1980': df_filtered[(df_filtered['year'] >= 1952) & (df_filtered['year'] <= 1980)]['co2'].sum(),
    '1980-2024': df_filtered[(df_filtered['year'] >= 1981) & (df_filtered['year'] <= 2024)]['co2'].sum()
}

#Conversion en DataFrame pour facilité de manipulation
periods_df = pd.DataFrame.from_dict(periods, orient='index', columns=['Émissions CO₂ (millions de tonnes)'])
periods_df = periods_df.reset_index().rename(columns={'index': 'Période'})

#Conversion en milliards de tonnes (Gt) pour meilleure lisibilité
periods_df['Émissions CO₂ (Gt)'] = periods_df['Émissions CO₂ (millions de tonnes)'] / 1000

#Création du graphique en barres
plt.figure(figsize=(12, 7))
bars = plt.bar(
    periods_df['Période'],
    periods_df['Émissions CO₂ (Gt)'],
    color=['#1f77b4', '#ff7f0e', '#d62728'],  # Bleu, orange, rouge
    width=0.6,
    edgecolor='black',
    linewidth=1
)

#Personnalisation du graphique
plt.title(
    'Émissions cumulées de CO₂ par période (1880-2024)',
    fontsize=16,
    pad=20,
    weight='bold'
)

plt.xlabel('Période', fontsize=14)
plt.ylabel('Émissions cumulées de CO2 (milliards de tonnes - Gt)', fontsize=14)

#Ajout des valeurs sur les barres
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2.,
        height,
        f'{height:.0f} Gt',
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold'
    )

#Améliorations visuelles
plt.ylim(0, periods_df['Émissions CO₂ (Gt)'].max() * 1.1)  # Marge en haut
plt.grid(axis='y', linestyle='--', alpha=0.7)

#Affichage
plt.tight_layout()
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Préparation des données (filtre des régions spéciales)
regions_speciales = [
    'Africa (GCP)', 'Asia (GCP)', 'Europe (GCP)', 'World',
    'International aviation', 'International shipping',
    'High-income countries', 'Low-income countries', 'OECD', 'Non-OECD'
]

#Filtre du DataFrame original pour exclure les régions spéciales
df_filtered = df[~df['country'].isin(regions_speciales)].copy()

# Ajout de la classification hémisphère à chaque pays
df_filtered['hemisphere'] = df_filtered['country'].map(
    lambda x: pays_mapping_strict.get(x, {}).get('hemisphere', None)
)

#Calcul des émissions totales par hémisphère (en tonnes)
emissions_par_hemisphere = df_filtered.groupby('hemisphere')['co2'].sum().reset_index()

# conversion en DataFrame propre
resultats = pd.DataFrame({
    'Hémisphère': ['Nord', 'Sud'],
    'Émissions totales (tonnes)': [
        emissions_par_hemisphere[emissions_par_hemisphere['hemisphere'] == 'Nord']['co2'].values[0],
        emissions_par_hemisphere[emissions_par_hemisphere['hemisphere'] == 'Sud']['co2'].values[0]
    ],
    'Émissions totales (Gt)': [
        emissions_par_hemisphere[emissions_par_hemisphere['hemisphere'] == 'Nord']['co2'].values[0] / 1e9,
        emissions_par_hemisphere[emissions_par_hemisphere['hemisphere'] == 'Sud']['co2'].values[0] / 1e9
    ]
})

# 4. Affichage des résultats sous forme de tableau
print("\n=== ÉMISSIONS TOTALES DE CO₂ PAR HÉMISPHÈRE (1880-2024) ===")
print(resultats.to_string(index=False))

# 5. Visualisation graphique (optionnel)
plt.figure(figsize=(10, 6))
bars = plt.bar(
    resultats['Hémisphère'],
    resultats['Émissions totales (Gt)'],
    color=['#1f77b4', '#d62728'],  # Bleu pour Nord, Rouge pour Sud
    width=0.5
)

# Ajout des valeurs sur les barres
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2.,
        height,
        f'{height:.1f} Gt',
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold'
    )

plt.title('Émissions cumulées de CO₂ par hémisphère (1880-2024)', fontsize=14, pad=20)
plt.ylabel('Émissions totales (milliards de tonnes - Gt)', fontsize=12)
plt.ylim(0, resultats['Émissions totales (Gt)'].max() * 1.15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 6. Calcul de la part relative
part_nord = resultats[resultats['Hémisphère'] == 'Nord']['Émissions totales (tonnes)'].values[0] / resultats['Émissions totales (tonnes)'].sum() * 100
part_sud = resultats[resultats['Hémisphère'] == 'Sud']['Émissions totales (tonnes)'].values[0] / resultats['Émissions totales (tonnes)'].sum() * 100

print(f"\n=== PART RELATIVE DES ÉMISSIONS ===")
print(f"Hémisphère Nord: {part_nord:.1f}%")
print(f"Hémisphère Sud: {part_sud:.1f}%")
print(f"Ratio Nord/Sud: {part_nord/part_sud:.1f}:1")


```
